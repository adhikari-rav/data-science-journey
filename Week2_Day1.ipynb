```python
import pandas as pd
import numpy as np

data = {
    "age": [25, 32, 28, 45, 23, 36, 29, 41, 27, 33],
    "gender": ["Male", "Female", "Female", "Male", "Non-binary",
               "Female", "Male", "Female", "Male", "Female"],
    "treatment": ["Yes", "No", "Yes", "Yes", "No",
                  "Yes", "No", "Yes", "No", "Yes"],
    "work_interfere": ["Often", "Rarely", "Often", "Sometimes", "Rarely",
                       "Sometimes", "Rarely", "Often", "Never", "Sometimes"],
    "mental_health_consequence": ["Yes", "No", "Maybe", "No", "Yes",
                                  "Maybe", "No", "Yes", "No", "Maybe"]
}

df = pd.DataFrame(data)

print("Dataset loaded!")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
messy_data = {
    "age": [25, 32, None, 45, 23, 36, 29, None, 27, 33],
    "gender": ["Male", "female", "Female", "MALE", "Non-binary",
               "Female", "male", "Female", "Male", "Female"],
    "treatment": ["Yes", "No", "Yes", "Yes", "No",
                  "Yes", "No", "Yes", "No", "Yes"]
}

df_messy = pd.DataFrame(messy_data)
print(df_messy)
```

```python
df_messy.info()
```

```python
print("Before cleaning:")
print(df_messy["age"].isnull().sum(), "missing values in age")

df_messy["age"] = df_messy["age"].fillna(df_messy["age"].mean())

print("After cleaning:")
print(df_messy["age"].isnull().sum(), "missing values in age")
```

```python
print("Before cleaning:")
print(df_messy["gender"].value_counts())

df_messy["gender"] = df_messy["gender"].str.lower()

print("\nAfter cleaning:")
print(df_messy["gender"].value_counts())
```

```python
df_messy["age"] = df_messy["age"].astype(int)
```
